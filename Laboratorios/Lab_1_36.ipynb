```python
# código

#lista con nombres de los archivos 
nombres = ["edad", "genero", "estado_civil", "escolaridad", "estrato", "region", "promedio"]

#función de lectura de archivos que guarda en una lista
def archivo_a_lista(directorio):
    archivo = open(directorio, "r")
    lista = archivo.readlines()
    datos = []
    for i in lista:
        datos.append(i.strip('\n'))
    return datos
        

#diccionario "data" donde se guardará toda la información en los archivos, llaves= los nombres"
data = {}

#lectura de todos los archivos en un ciclo que guarda la información en "data"
for i in range(len(nombres)):
    data[nombres[i]] = archivo_a_lista("./Archivos/"+nombres[i]+".txt")

#variables númericas promedio y edad
data["promedio"] = list(map(float, data["promedio"]))
data["edad"] = list(map(int, data["edad"]))
data["estrato"] = list(map(int, data["estrato"]))
```

```python
# código

# importación de paquetes necesarios
import matplotlib.pyplot as plt

# función que calcula la mediana de una lista de datos numéricos

def mediana(lista):
    lista2 = lista.copy()
    lista2.sort()
    if len(lista)%2 > 0:
        med = lista2[int(len(lista)/2)]
    else:
        ini = max([0, len(lista)/2-1])
        fin = min([len(lista), len(lista)/2])
        med = (lista2[int(fin)] + lista2[int(ini)])/2
    return med


# función que calcula la desviación de una lista de datos numéricos

def desviacion(lista):
    promedio = sum(lista)/len(lista)
    dif_c = []
    for i in range(len(lista)):
        dif_c.append((lista[i]-promedio)**2)
    desv = pow(sum(dif_c)/(len(dif_c)-1), 0.5)
    return desv

#función de estadísticas básicas
def estad(lista, factor=[]):
    if factor:
        if len(lista) == len(factor):
            uni_factor = list(dict.fromkeys(factor))
            for i in range(len(uni_factor)):
                x = []
                for j in range(len(lista)):
                    if factor[j] == uni_factor[i]:
                        x.append(lista[j])
                print(uni_factor[i])
                print("n:", len(x))
                print("Promedio:", sum(x)/len(x))
                print("Desviación Estándar:", desviacion(x))
                print("Mínimo:", min(x))
                print("Mediana:", mediana(x))
                print("Máximo:", max(x))
                print("")
        else: print("La lista y el factor deben tener el mismo tamaño")
    else: 
        print("n:", len(lista))
        print("Promedio:", sum(lista)/len(lista))
        print("Desviación Estándar:", desviacion(lista))
        print("Mínimo:", min(lista))
        print("Mediana:", mediana(lista))
        print("Máximo:", max(lista))
        print("")

#función para graficar frecuencias
def grafica(lista, titulo=""):
    eje_x = list(dict.fromkeys(lista))
    eje_y = []
    for i in range(len(eje_x)):
        cont = 0
        for j in range(len(lista)):
            if eje_x[i] == lista[j]:
                cont += 1
        eje_y.append(cont)
    plt.bar(eje_x, eje_y)
    plt.ylabel('Frecuencia')
    if titulo:
        plt.title('Gráfica de Frecuencia de '+ titulo)
    else:
        plt.title('Gráfica de Frecuencia')
    plt.show()


###################### RESULTADOS ######################

#Estadísticas básicas de la edad
print("Estadísticas de Edad")
estad(data["edad"])

#Estadísticas básicas de la calificación promedio por las demás variables
for i in ["genero", "estado_civil", "escolaridad", "estrato", "region"]:
    print("Estadísticas de calificación por", i)
    estad(data["promedio"], data[i])
    print("")

#Gráficas de barras de las variables cualitativas
for i in ["genero", "estado_civil", "escolaridad", "estrato", "region"]:
    print("")
    grafica(data[i], i)
    print("")
```

```python
# código

   
#función que calcula los n >= 0 de una lista de acuerdo con un factor, el resultado es un diccionario
#con llaves cada nivel del factor y valores los n >= 0 por factor o los n >= 0 de la lista si no hay factor    
def cuenta_positiva(lista, factor=[]):
    cuenta = []
    if factor:
        if len(lista) == len(factor):
            uni_factor = list(dict.fromkeys(factor))
            for i in range(len(uni_factor)):
                x = []
                for j in range(len(lista)):
                    if factor[j] == uni_factor[i] and lista[j] >= 0:
                        x.append(lista[j])
                cuenta.append(len(x))
            resultado = dict(zip(uni_factor, cuenta))
            return resultado
        else: print("La lista y el factor deben tener el mismo tamaño")
    else: 
        return len(lista)

#función que arroja los indices donde se encuentra un estrato dado (nivel)
def selec_estrato(factor, nivel):
    indices = []
    for i in range(len(factor)):
        if factor[i] == nivel:
            indices.append(i)
    return indices

#función que arroja los indices de los n valores máximos dentro de una lista, dado un factor (p.ej. estrato), un nivel, n: "tope"
def selec_max_factor(lista, factor, nivel, tope):
    lista_cal = []
    i = 1
    indices = []
    niveles = selec_estrato(factor, nivel)
    for j in niveles:
        lista_cal.append(lista[j])
    while i <= tope:
        if i > len(niveles): break
        if max(lista_cal) < 0: break
        x = lista_cal.index(max(lista_cal))
        indices.append(niveles[x])
        lista_cal[x] = -1
        i += 1
    return indices
  
#función que selecciona de acuerdo con el método 1, arroja los indices de los n valores máximos dentro de una lista 
#por factor (estrato), dado un porcentaje por grupo (2%) y un número de becas máximo
def indices_selec1(lista, factor, porcentaje, num_becas):
    lista_cal = lista.copy()
    uni_factor = list(dict.fromkeys(factor))
    uni_factor.sort()
    becas = 0
    indices = []
    while becas < num_becas:
        for i in range(len(uni_factor)):
            cont = 0
            tope_factor = int(cuenta_positiva(lista_cal, factor)[uni_factor[i]]*porcentaje)
            if tope_factor > (num_becas - becas):
                tope = num_becas - becas
            else: tope = tope_factor
            indices = indices + selec_max_factor(lista_cal, factor, uni_factor[i], tope)
            for j in selec_max_factor(lista_cal, factor, uni_factor[i], tope):
                lista_cal[j] = -1
            becas = len(indices)
    return indices 

#ejemplo selección método 1 por estrato y 2% de personas con promedio académico superior, se toma como ejemplo un 
#número de becas de 30, los indices se guardan en la lista "seleccionados1"
seleccionados1 = indices_selec1(data["promedio"], data["estrato"], 0.02, 30)

print("Individuos seleccionados grupo 1")
print(seleccionados1)

#guardando el grupo 1 en un diccionario nuevo
est_selec = []
prom_selec = []
reg_selec = []
edad_selec = []
est_c_selec = []
escol_selec = []
gene_selec = []

for i in range(len(seleccionados1)):
    for j in range(len(data["promedio"])):
        if seleccionados1[i] == j:
            est_selec.append(data["estrato"][j])
            prom_selec.append(data["promedio"][j])
            reg_selec.append(data["region"][j])
            edad_selec.append(data["edad"][j])
            est_c_selec.append(data["estado_civil"][j])
            escol_selec.append(data["escolaridad"][j])
            gene_selec.append(data["genero"][j])

grupo1 = dict(zip(nombres, [edad_selec, gene_selec, est_c_selec, escol_selec, est_selec, reg_selec, prom_selec]))
 
```
