### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado.
2. Número de niños en el hogar.
3. Número de adolescentes en el hogar.

```python
# YOUR CODE HERE
def filtrar(Educacion, ninos, adolescentes):
    datos = mercadeo
    if  Educacion != "Todos" and ninos != "Todos" and adolescentes != "Todos":
        display(datos[(datos["Education"] == Educacion) & (datos["Kidhome"] == ninos) & (datos["Teenhome"] == adolescentes)])
    elif Educacion == "Todos":
        display(datos[(datos["Kidhome"] == ninos) & (datos["Teenhome"] == adolescentes)])
    elif ninos == "Todos":
        display(datos[(datos["Education"] == Educacion) & (datos["Teenhome"] == adolescentes)])
    elif adolescentes == "Todos":
        display(datos[(datos["Education"] == Educacion) & (datos["Kidhome"] == ninos)])
    else: 
        display(datos)

interactive_plot = interactive(
                                filtrar, 
                                Educacion = mercadeo['Education'].unique().tolist(),
                                ninos = mercadeo['Kidhome'].unique().tolist(),
                                adolescentes = mercadeo['Teenhome'].unique().tolist()
                              )
display(interactive_plot)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
# YOUR CODE HERE
mercadeo_bying = mercadeo.sort_values('Income')
gasto = widgets.Dropdown(
                        description = 'Variable de gasto:',
                        options = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'],
                        style = {'description_width': 'initial'}       # Mostrar el nombre completo.
                       )

fig1 = go.FigureWidget(
                        data = [go.Scatter()],  
    
                        layout = go.Layout(
                                            title = 'Ingreso vs Gasto',
                                            xaxis_title = 'Ingreso',
                                            yaxis_title = 'Gasto',
                                          ))
fig2 = go.FigureWidget(
                        data =[go.Box(name = "Ingreso")],
                        layout = go.Layout(
                                            title = 'Boxplot del Ingreso'))
fig3 = go.FigureWidget(
                        data =[go.Box(name = "Gasto")],
                        layout = go.Layout(
                                            title = 'Boxplot del Gasto'))
def filtrar():
    df_temporal = mercadeo_bying.loc[: , ['Income', gasto.value]]
    fig1.data[0].x = df_temporal['Income']
    fig1.data[0].y = df_temporal[gasto.value]
    fig2.data[0].y = df_temporal['Income']
    fig3.data[0].y = df_temporal[gasto.value]
def respuesta(change):
    filtrar()
gasto.observe(respuesta)
filtrar()
display(gasto, fig1, fig3, fig2)
```
